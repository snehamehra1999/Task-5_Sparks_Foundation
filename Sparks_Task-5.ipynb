```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import plotly_express as px
%matplotlib inline
```

```python
df = pd.read_csv("globalterrorism.csv",low_memory = False, encoding = "latin-1")
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#dropping which has all the null values
df.dropna(axis=1,how="all")
```

```python
#attacks in regions according to the year
pd.crosstab(df.iyear,df.region_txt).plot(color =sbn.color_palette("Paired",12))
fig =plt.gcf()
fig.set_size_inches(18,6)
```

```python
#Number of terror Activities
plt.subplots(figsize =(20,10))
sbn.countplot("iyear", data = df, palette = "tab10", edgecolor = sbn.color_palette("hls",8))
plt.xticks(rotation = 90)
plt.title("Number Of Terror Activities")
```

```python
ay = df.iyear.value_counts().to_dict()
rate = ((ay[2017]-ay[1970])/ay[2017])*100
```

```python
print("Number of Atacks in the Year 1970 is:", ay[1970])
print("Number of Atacks in the Year 2017 is:", ay[2017])
```

```python
pd.crosstab(df.region_txt,df.attacktype1_txt).plot.barh(stacked =True,width=1,color
                =sbn.color_palette("Paired",9))
sv = plt.gcf()
sv.set_size_inches(12,8)
plt.show()
```

```python
plt.figure(figsize= (20,10))
sbn.countplot(df["targtype1_txt"],order = df["targtype1_txt"].value_counts().index, palette = "Paired")
plt.xticks(rotation = 90)
```

```python
#Countries with the highest number of attacks
fig,axes = plt.subplots(figsize = (20,10), nrows =1, ncols =1)
sbn.barplot(x =df["country_txt"].value_counts()[:20].values, y =df["country_txt"]
           .value_counts()[:20].index,palette = "mako")
```

```python
#Regions with Most Under Attack
fig,axes = plt.subplots(figsize = (20,10), nrows =1, ncols =1)
sbn.barplot(x =df["region_txt"].value_counts()[:20].values, y =df["region_txt"]
           .value_counts()[:20].index,palette = "rocket")
```
